```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import db
import pandas as pd

# Initialize Firebase
cred = credentials.Certificate('../../service-account.json')
firebase_admin.initialize_app(cred, {
    'databaseURL': 'https://codeblue-482a2-default-rtdb.firebaseio.com/'
})
```

```python

# Get a reference to the "hospitals" collection
hospitals_ref = db.reference('hospitals')

# Push each row as a new document with a unique name under "hospitals" collection
for i, row in df.iterrows():
    # Generate a new unique key for the document
    new_doc_ref = hospitals_ref.push()
    # Set the values of the document
    new_doc_ref.set({
        'name': row['name'],
        'lat': row['lat'],
        'lon': row['lon'],
        'affiliation': row['affiliation'],
        'ambulances': row['ambulances'],
        'available': row['available']
    })
```

```python
import firebase_admin
from firebase_admin import credentials
from firebase_admin import firestore
import pandas as pd

# Initialize Firestore
cred = credentials.Certificate('../../service-account.json')
firebase_admin.initialize_app(cred)
db = firestore.client()
```

```python
# Assuming your DataFrame is named "df"
for _, row in df.iterrows():
    data = {
        'name': row['name'],
        'lat': row['lat'],
        'lon': row['lon'],
        'affiliation': row['affiliation'],
        'ambulances': row['ambulances'],
        'available': row['available']
    }
    # Create a new document with an auto-generated ID using push()
    doc_ref = db.collection('hospitals').document()
    doc_ref.set(data)
```
